```python
# car data 정보 확인
car.info()
car.info
```

```python
# 이상치 제거해줘야함
# print(car["normalized-losses"]!="?")
a = car[car["normalized-losses"]!="?"]        # 정상적인놈들은 a에 저장
b = (a["normalized-losses"].astype(int)).mean()  #정상적인놈들 
car["normalized-losses"] = car["normalized-losses"].replace("?",b).astype(int)
```

```python
a= car[car["body-style"] == "sedan"]
a["num-of-doors"].value_counts()
```

세단은 문 개수가 거의다 4개구나. 
?는 4개로 바꾸자

```python
# 문 개수 결측치 제거
a = car["num-of-doors"].map({"two":2, "four":4, "?" :4})
car["num-of-doors"] = a
```

```python
# 가격 결측치 제거
a= car[car["price"] != "?"]
b = (a["price"].astype(int)).mean()
car["price"] = car["price"].replace("?",b).astype(int)
```

결측치에 차 가격을 평균내도되는건가... 싶지만

```python
# 마력 결측치 제거
a= car[car["horsepower"] != "?"]
b = (a["horsepower"].astype(int)).mean()
car["horwepower"] = car["horsepower"].replace("?",b).astype(int)
```

```python
# 보어는 피스톤의 직경을 말함 // 결측치 제거
a= car[car["bore"] != "?"]
b = (a["bore"].astype(float)).mean()
car["bore"] = car["bore"].replace("?",b).astype(float)
```

#### *엔진블록의 실린더 내부 지름을 깎은걸 보링이라 한다.

```python
# 스트로크는 피스톤이 상하운동 할 수 있는 거리를 말함. // 결측치 제거
a= car[car["stroke"] != "?"]
b = (a["stroke"].astype(float)).mean()
car["stroke"] = car["stroke"].replace("?",b).astype(float)
```

```python
# 최대 알피엠 결측치 제거
a= car[car["peak-rpm"] != "?"]
b = (a["peak-rpm"].astype(float)).mean()
car["peak-rpm"] = car["peak-rpm"].replace("?",b).astype(float)
```

```python
a = car["num-of-cylinders"].map({"four" :4, "five":5, "six":6, "?" :4})
car["num-of-cylinders"] = a
```

```python
car.describe().T
```

```python
f, ax = plt.subplots(1,2,figsize = (18,8))
car["make"].value_counts().plot.pie(autopct = "%1.1f%%", ax = ax[0], shadow = True)
ax[0].set_title("Make of car")

sns.countplot("make", data = car, ax = ax[1], 
              order = car["make"].value_counts().index)
ax[1].set_title("Make of car")

plt.show()
```

- 빈도표를 plot해주는 crosstab 함수

```python
f,ax = plt.subplots(1,2,figsize = (18,8))
car["fuel-type"].value_counts().plot.pie(autopct = "%1.1f%%",
                                         explode = [0,0.1], # 파이 떨어뜨리는 정도
                                         ax = ax[0])
ax[0].set_title("Fuel Type")   
ax[0].set_ylabel("Count")

sns.countplot("fuel-type",data = car, ax = ax[1], 
               order = car["fuel-type"].value_counts().index)
ax[1].set_title("Fuel Type")

plt.show()
```

```python
f,ax = plt.subplots(1,2,figsize = (18,8))
car["aspiration"].value_counts().plot.pie(autopct = "%1.1f%%",
                                         explode = [0,0.1], # 파이 떨어뜨리는 정도
                                         ax = ax[0])
ax[0].set_title("aspiration type")   
ax[0].set_ylabel("Count")

sns.countplot("aspiration",data = car, ax = ax[1], 
               order = car["aspiration"].value_counts().index)
ax[1].set_title("aspiration type")

plt.show()
```
